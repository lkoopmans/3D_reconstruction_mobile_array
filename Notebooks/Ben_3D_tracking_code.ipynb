```python
dataout = pd.DataFrame(columns=['track', 'ID', 'X','Y','Z','FRAME_IDX','errors'])


os.chdir("/Users/benmartin/")        

dat = pd.read_csv('Google Drive/Projects/BonaireTracking/clips_all_cam_for_trackings/Sync_times.csv')

#dat = dat[dat["Video"] != 451]
#dat = dat[dat["Video"] != 455]
#dat = dat[dat["Video"] != 456]
#dat = dat[dat["Video"] != 283]
#dat = dat[dat["Video"] != 243]

dat = dat[dat["Video"] != 251]
dat = dat[dat["Video"] != 334]
dat = dat[dat["Video"] != 204]
data = dat[dat["Video"] != 431]
data = data.reset_index()

colmap_location= pd.read_csv('Google Drive/Projects/BonaireTracking/clips_all_cam_for_trackings/best_colmap_model_dir.csv')



os.chdir('/Users/benmartin/Google Drive/Projects/BonaireTracking/clips_all_cam_for_trackings')
for i in range(np.size(data.Video)):
    os.chdir(str(data.Video[i]))
    print(str(data.Video[i]))
    x=colmap_location[colmap_location["tracks"]==data.Video[i]]
    x=x["out"].iloc[0]
    y='colmap/'
    pathname=y+x
    scene = mvt.Scene(model_path=pathname,
                  tracks_path='h5_tracking_files',
                  fisheye=True,
                  verbose=False)
    scene.get_cameras()
    scene.interpolate_cameras()
    scene.triangulate_multiview_tracks()
    camera_ids = [1, 3]
    world_distance = 1
    n_frames= np.size(scene.tracks_triangulated['FRAME_IDX'])
    reconstruction_errors = scene.scale(camera_ids, world_distance,n_frames)
    reconstruction_errors=reconstruction_errors[0:(n_frames)]
    #scene.rotate()
    IDs = scene.tracks_triangulated['IDENTITIES']
    for j in IDs:
        temp = pd.DataFrame.from_dict(scene.tracks_triangulated[str(j)])
        temp.insert(0,"ID",j)
        temp.insert(0,"track",data.Video[i])
        temp.insert(0,"errors",reconstruction_errors)
        dataout = pd.concat([dataout,temp])
    os.chdir('..')
dataout.to_csv("3Dtracks_first_cut_bundle_not_rotated_3.csv",index=False)
```
